**Выводы**: 
отобранные в 2021 году образцы коронавируса распадаются на несколько хорошо различимых клад. Это свидетельствует о том, что на тот момент в регионе присутствовало значительное эволюционное разнообразие штаммов. Те образцы, у которых обнаружены одинаковые недавно возникшие мутации, оказываются на соседних ветвях эволюционного дерева.

Берем 3 последовательности Spike белка из других штаммов:

Штамм Beta (ЮАР) — Accession ID: MZ314997

Штамм Gamma (Бразилия) — Accession ID: MZ477858

Штамм Delta (Индия) — Accession ID: MZ359841

```python

original_file = 'S_prot_sequences.fasta'


new_files = [
    'SARS-COV/South_Africa_Beta.fasta',
    'SARS-COV/Brazil_Gamma.fasta',
    'SARS-COV/India_Delta.fasta'
]


with open('combined_S_proteins.fasta', 'w') as outfile:
    with open(original_file, 'r') as infile:
        outfile.write(infile.read() + '\n')
    
    for fname in new_files:
        with open(fname, 'r') as infile:
            outfile.write(infile.read() + '\n')

```
